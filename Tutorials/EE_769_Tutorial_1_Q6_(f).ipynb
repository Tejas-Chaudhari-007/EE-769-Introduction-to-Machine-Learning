```
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter

# Gaussian function
def Gaussian_PDF(x, mean, std_dev):
  Gaussian_pdf = 1/np.sqrt(2*np.pi*std_dev**2)*np.exp(-(x-mean)**2/2/std_dev**2)
  return Gaussian_pdf

# Define x and y variables
x = np.arange(-5.5, 5.5, 0.1)
y = np.arange(-5.5, 5.5, 0.1)

# Call the function
pdf_x = Gaussian_PDF(x, 0, 1)
pdf_y = Gaussian_PDF(y, 0, 2)

# Plot the graphs
plt.plot(x, pdf_x, color='red', linestyle='dashed', linewidth=2, label='PDF of X')
plt.plot(y, pdf_y, color='green', linewidth=2, label='PDF of Y')
plt.legend()
plt.xlim(-5.5, 5.5)
plt.xticks(np.arange(-5, 6, step=1))
plt.xlabel('Random variable')
plt.ylabel('PDF')
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))  # plt.gca() returns the axes instances
```

### (ii) From the above graph we can see that area under the curve after 3 in PDF of Y is higher than that of X. Hence probability that the value is greater than 3 is higher in Y.
